```python
# Agrupamos por Emissora e Audiencia Domiciliar
df_Ibope = df.groupby('Emissora')['AudienciaDomiciliar'].sum().reset_index().sort_values(by = 'AudienciaDomiciliar',
                                                                                    ascending = False).head()
```

```python
# Função para converter os dados em valor absoluto
def autopct_format(values): 
    def my_format(pct): 
        total = sum(values) 
        val = int(round(pct * total / 100.0))
        return ' $ {v:d}'.format(v = val)
    return my_format
```

```python
df_Ibope
```

```python
# Plot
# Tamanho da figura
plt.figure(figsize = (16, 6))

# Gráfico de pizza
plt.pie(df_Ibope['AudienciaDomiciliar'], 
        labels = df_Ibope['Emissora'],
        autopct = autopct_format(df_Ibope['AudienciaDomiciliar']),
        startangle = 90)

# Limpa o círculo central
centre_circle = plt.Circle((0, 0), 0.82, fc = 'white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Labels e anotações
plt.annotate(text = 'Audiências: ' + '$ ' + str(int(sum(df_Ibope['AudienciaDomiciliar']))), xy = (-0.25, 0))
plt.title('Total de Audiências Domiciliar Por Emissora')
plt.show()
```
